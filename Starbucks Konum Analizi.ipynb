```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import LeavePOut
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
#Regresyon ve model i√ßin k√ºt√ºphanelerin dahil edilmesi
import math as math
import statistics
```

```python
consdata['Stores_count_log'] = consdata['Stores_count'].astype(float).apply(lambda x: math.log(x)) #math.log ile sayƒ±nƒ±n tabana g√∂re logaritmasƒ±na d√∂nd√ºrme
consdata['Total_log'] = consdata['Total'].astype(float).apply(lambda x: math.log(x))
consdata
```

G√∂receli olarak √ßok k√º√ß√ºk bir veri seti olduƒüu g√∂r√ºld√ºƒü√º i√ßin, sadece 26 satƒ±r ve hatta bazƒ±larƒ± NaN deƒüerde, √ßok az g√∂zlem i√ßin doƒürusal bir √ßizgi uydurmada sorun ya≈üanabilir. Leave P Out Cross Vaildation kullanarak Lineer Regresyon modelimizi geli≈ütirmeye √ßalƒ±≈üƒ±lacaktƒ±r. Bu y√∂ntem, hesaplama a√ßƒ±sƒ±ndan aƒüƒ±r olmasƒ±na raƒümen, buradaki gibi √ßok k√º√ß√ºk veri k√ºmeleri i√ßin katlanƒ±labilir olmasƒ± ve ortalamasƒ±nƒ± alabilinecek bir√ßok puan hesaplamasƒ± vermesi nedeniyle se√ßilmi≈ütir.

Maƒüazalar ve Ki≈üi Ba≈üƒ± Kahve T√ºketiminin Regresyon Analizi

```python

x = consdata['PerCapita'].loc[(consdata['PerCapita'].isna() == False) & (consdata['Stores_count_log'].isna() == False)].reset_index(drop=True)
y = consdata['Stores_count_log'].loc[(consdata['PerCapita'].isna() == False) & (consdata['Stores_count_log'].isna() == False)].reset_index(drop=True)
```

```python
#Linear Regression
lpoparam = 4
lpo = LeavePOut(lpoparam)

r2list = []
for train_index, test_index in lpo.split(x):
    x_train, x_test = np.array(x[train_index]).reshape(-1,1), np.array(x[test_index]).reshape(-1,1)
    y_train, y_test = np.array(y[train_index]).reshape(-1,1), np.array(y[test_index]).reshape(-1,1)
    reg = LinearRegression()
    reg.fit(x_train,y_train)
    y_pred = reg.predict(x_test)
    r2list.append(r2_score(y_test, y_pred))
    
print(f'Leave P-Out of {lpoparam} with R2 mean of {statistics.mean(r2list)} with maximum R2 of {max(r2list)}')
```

```python
x_train, x_test, y_train, y_test = train_test_split( x , y , test_size=0.3 , random_state=42)
reg = LinearRegression()
reg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred = reg.predict(np.array(x_test).reshape(-1,1))
r2_score(np.array(y_test).reshape(-1,1), y_pred) 
StoVsCapParams = reg.coef_
```

```python
#Regression Tree
x_train, x_test, y_train, y_test = train_test_split( x , y , test_size=0.3 , random_state=42)
reg = LinearRegression()
reg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred = reg.predict(np.array(x_test).reshape(-1,1))
r2_score(np.array(y_test).reshape(-1,1), y_pred)
```

Maƒüazalar ve Toplam Kahve T√ºketiminin Regresyon Analizi


```python
x = consdata['Total_log'].loc[(consdata['Total_log'].isna() == False) & (consdata['Stores_count_log'].isna() == False)].reset_index(drop=True)
y = consdata['Stores_count_log'].loc[(consdata['Total_log'].isna() == False) & (consdata['Stores_count_log'].isna() == False)].reset_index(drop=True)
```

```python
#leave p out cross validation, √ßapraz doƒürulama
lpoparam = 4
x = consdata['Total_log'].loc[consdata['Total_log'].isna() == False].reset_index(drop=True)
y = consdata['Stores_count_log'].loc[consdata['Total_log'].isna() == False].reset_index(drop=True)
lpo = LeavePOut(lpoparam)

r2list = []
for train_index, test_index in lpo.split(x):
    x_train, x_test = np.array(x[train_index]).reshape(-1,1), np.array(x[test_index]).reshape(-1,1)
    y_train, y_test = np.array(y[train_index]).reshape(-1,1), np.array(y[test_index]).reshape(-1,1)
    reg = LinearRegression()
    reg.fit(x_train,y_train)
    y_pred = reg.predict(x_test)
    r2list.append(r2_score(y_test, y_pred))
    
print(f'Leave P-Out of {lpoparam} with R2 mean of {statistics.mean(r2list)} with maximum R2 of {max(r2list)}')
```

```python
#Normal
x_train, x_test, y_train, y_test = train_test_split( x , y , test_size=0.3 , random_state=42)
reg = LinearRegression()
reg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred = reg.predict(np.array(x_test).reshape(-1,1))
r2_score(np.array(y_test).reshape(-1,1), y_pred)
StoVsTotParams = reg.coef_
```

```python
#Decision Tree Regression
x_train, x_test, y_train, y_test = train_test_split( x , y , test_size=0.3 , random_state=42)
reg = DecisionTreeRegressor(random_state=42)
reg.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
y_pred = reg.predict(np.array(x_test).reshape(-1,1))
r2_score(np.array(y_test).reshape(-1,1), y_pred)
```

Bir kez daha, Leave P Out y√∂ntemi bize k√∂t√º sonu√ßlar verirken, karar aƒüacƒ± 0,69 R2 ile nispeten iyi sonu√ßlar vermi≈ütir, ancak R2'yi 0,72 veren lineer regresyona kƒ±yasla hala d√º≈ü√ºk, bu da iyi bir √∂ng√∂r√ºye i≈üaret eder.

## Maƒüaza Tahmini 


Lineer regresyonun modele uygunluƒüu tespit edildikten sonra gelecekteki maƒüaza a√ßƒ±lƒ±≈ülarƒ± ve geni≈ülemeler i√ßin potansiyel yerler tahmin etmeye √ßalƒ±≈üƒ±lacak. Bunun i√ßin, train veri setinin, toplam t√ºketimine kar≈üƒ± maƒüazalarƒ±n grafiksel diyagramƒ±ndan olu≈üturulan grafiƒüi kullanacaktƒ±r. Sklearn veri seti i√ßin uygun hat kat sayƒ±lar saƒülamadƒ±ƒüƒ±ndan dolayƒ± doƒürusal uyumu √ßok daha iyi olan tahminci kullanƒ±lacaktƒ±r.  
Maƒüazalarƒ± 2 kategoriye ayrƒ±lacaktƒ±r; 
  - √áizginin altƒ±nda kalan alanda, bir √ºlkenin maƒüazalarla 'az n√ºfuslu' olduƒüu, o √ºlkenin toplam kahve t√ºketimini kar≈üƒ±layacak yeterli maƒüaza olup olmadƒ±ƒüƒ± ve daha fazla geni≈üleme i√ßin yer olup olmadƒ±ƒüƒ± anlamƒ±na gelir.
  - √áizginin √ºzerindeki √ºlkeler, √ºlkenin maƒüazalarla 'a≈üƒ±rƒ± kalabalƒ±k' olduƒüu veya o √ºlkenin toplam kahve t√ºketimini kar≈üƒ±lamak i√ßin √ßok fazla maƒüaza olduƒüu anlamƒ±na gelir.

## Hyperparameters Tuning

```python
# Random Forest'taki tree sayƒ±sƒ±
n_estimators = [int(x) for x in np.linspace(start = 50, stop = 500, num = 4)]
# Her b√∂lmede dikkate alƒ±nmasƒ± gereken √∂zellik sayƒ±sƒ±
max_features = ['auto', 'sqrt']
# Aƒüa√ßtaki maksimum seviye sayƒ±sƒ±
max_depth = [int(x) for x in np.linspace(10, 110, num = 4)]
max_depth.append(None)
# Her aƒüacƒ± eƒüitmek i√ßin √∂rnek se√ßme y√∂ntemi
bootstrap = [True, False]# random gridleri olu≈üturma
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'bootstrap': bootstrap}

print(random_grid)
```

```python
# En iyi hiper parametreleri aramak i√ßin rastgele ƒ±zgarayƒ± kullanƒ±n
# √ñnce ayarlanacak temel modeli olu≈üturun
rf = RandomForestClassifier()
# 3 kat √ßapraz doƒürulama kullanarak parametrelerin rastgele aranmasƒ±,
# 100 farklƒ± kombinasyonda arama yapƒ±lƒ±p mevcut t√ºm √ßekirdeklerin kullanƒ±lmasƒ±
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=10, random_state=42, n_jobs = -1)
# random search modelinin fit edilmesi
rf_random.fit(train_X, train_y)
```
